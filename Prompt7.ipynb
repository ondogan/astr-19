```python
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
#Here I will define the given interval
n=100
x = np.linspace(0,1,n)
```

```python
#Here I will define my function 
def func_1(x):
    return np.exp(x)
```

```python
#Here set the y to my function and plot and save the figure

y = func_1(x)
plt.plot(x,y)
plt.xlabel("Time[Miliseconds]")
plt.ylabel("Awesomeness")


plt.savefig("Prompt7.png",bbox_inches='tight',dpi=400)
```
