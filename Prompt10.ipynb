```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 1000
x = np.random.randn(n)

histmin = np.floor(min(x))
histmax = np.ceil(max(x))

bins = np.linspace(histmin,histmax,100)
plt.hist(x,bins=bins,edgecolor="red")
plt.xlabel("class intervals")
plt.ylabel("Density Scale")
plt.savefig("prompt10.pdf", dpi=300,bbox_inches="tight")
plt.show()
```
