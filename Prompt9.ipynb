```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# vs matplotlib qt
```

The Assignment is the following:

Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. 
Histogram the random numbers into 100 bins,
and plot the histogram. 
Label your axes and save the figure as a PDF.


```python
#Generate 1000 uniformly distributed(same increments) data between 0 and 1
n=1000

x = np.random.uniform(0,1,n)

#create the histogram with 100 bins 

#create 100 class intervals, each class interval will get some number of entries coming from x distrubution
# the area of the bin will be the percantage of the numbers within that interval



histmin = np.floor(min(x))
histmax = np.ceil(max(x))

bins = np.linspace(histmin,histmax,100)

plt.hist(x,bins=bins,edgecolor="red")
plt.xlabel("x")
plt.ylabel("Density Scale")
plt.savefig("prompt9.pdf", dpi=300,bbox_inches="tight")
plt.show()
```
